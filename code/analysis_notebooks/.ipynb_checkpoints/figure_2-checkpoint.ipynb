## Meridional and temporal fit to zonally averaged TAO and NEMO data

```python
import sys
sys.path.append('../')
import matplotlib.pyplot as plt
import numpy as np
import xarray as xr
from numpy import pi
import scipy.io as sio
from tqdm import tqdm
import matplotlib.colors as Colors
%matplotlib inline 
import warnings
from tools.transform_tools import *
from tools.data_processing_tools import *
from tools.theoretical_tools import *

warnings.filterwarnings('ignore')
plt.rcParams.update({'font.size': 14, 'font.size': 14})
plt.rcParams['figure.figsize'] = (20, 10)
plt.rcParams['text.usetex'] = False

%load_ext autoreload
%autoreload 2
```

```python
smooth_N = 21 # Rolling mean length for first high pass filter
NSR = 35 # Noise to signal ratio for least squares fit
nmodes=6 # Number of meridional modes to fit to
cutoff = 20 # Cutoff period for second high pass filter
```

```python
# Load in full TAO dynamic height data 
t_TAO_full, lat_TAO, lon_TAO, lon_TAO_midpoints, D_TAO_full, ds_TAO_full = load_TAO(NEMO_year=False)

# Load in full NEMO dynamic height data
t, lat_NEMO, lon_NEMO, D_NEMO_full, ds_NEMO_full = load_NEMO(daily_mean=True,lons='default',lats='default',lon_lims=[lon_TAO_midpoints[0],lon_TAO_midpoints[-1]])

# Load in year of TAO dynamic height data 
t, lat_TAO, lon_TAO, lon_TAO_midpoints, D_TAO_year, ds_TAO_year = load_TAO(NEMO_year=True)

# Load in sampled NEMO dynamic height data
t, lat_TAO, lon_TAO, D_NEMO_sampled, ds_NEMO_sampled = load_NEMO(daily_mean=True,lons=lon_TAO,lats=lat_TAO)

# Add in gaps from TAO data to dynamic height NEMO data:
D_NEMO_sampled = np.where(np.isnan(D_TAO_year),np.nan,D_NEMO_sampled)
```

```python
# Get the theoretical frequencies
freqbc1, freqbc2, c1, c2, lon_freqs = find_predicted_freqs(nmodes,lon_lims=[lon_TAO_midpoints[0],lon_TAO_midpoints[-1]],lat_lims=[-12,12],average_lon = True)
```

```python
# Load in meridional modes. Use the first baroclinic mode.
P_modes_NEMO = calc_meridional_modes(lat_NEMO,N=nmodes,cm=c1)[1]
P_modes_TAO = calc_meridional_modes(lat_TAO,N=nmodes,cm=c1)[1]
```

```python
# Do high pass filters (two stage approach - first remove a running mean to remove endpoint issues)

D_NEMO_full_hf1 = D_NEMO_full - smooth(D_NEMO_full,smooth_N)
D_NEMO_sampled_hf1 = D_NEMO_sampled - smooth(D_NEMO_sampled,smooth_N)
D_TAO_year_hf1 = D_TAO_year - smooth(D_TAO_year,smooth_N)

D_NEMO_full_lf = least_squares_spectrum_t_multi(D_NEMO_full_hf1, t, min_period=2, max_period=370, NSR=NSR, reconstruct_min_period = cutoff)[2]
D_NEMO_full_hf = D_NEMO_full_hf1 - D_NEMO_full_lf
D_NEMO_full_hf -= np.nanmean(D_NEMO_full_hf,axis=0)

D_NEMO_sampled_lf = least_squares_spectrum_t_multi(D_NEMO_sampled_hf1, t, min_period=2, max_period=370, NSR=NSR, reconstruct_min_period = cutoff)[2]
D_NEMO_sampled_hf = D_NEMO_sampled_hf1 - D_NEMO_sampled_lf
D_NEMO_sampled_hf -= np.nanmean(D_NEMO_sampled_hf,axis=0)

D_TAO_year_lf = least_squares_spectrum_t_multi(D_TAO_year_hf1, t, min_period=2, max_period=370, NSR=NSR, reconstruct_min_period = cutoff)[2]
D_TAO_year_hf = D_TAO_year_hf1 - D_TAO_year_lf
D_TAO_year_hf -= np.nanmean(D_TAO_year_hf,axis=0)

```

```python
# For the full TAO data, we need a windowed approach. Let's first create the windowed data, in yearly (370 day) intervals
window_width=370
Nsamples = D_TAO_full.shape[0]
window_starts = np.arange(0,Nsamples-window_width,window_width)
window_mid = window_starts+window_width/2
Nwindows = window_starts.shape[0]
D_TAO_windowed = np.zeros((window_width,D_TAO_full.shape[1],D_TAO_full.shape[2],Nwindows))
D_TAO_full_lf = np.zeros_like(D_TAO_windowed)
D_TAO_full_hf1 = np.zeros_like(D_TAO_windowed)
for iw in range(Nwindows):
    D_TAO_windowed[:,:,:,iw] = D_TAO_full[window_starts[iw]:window_starts[iw]+window_width]
    D_TAO_full_hf1[:,:,:,iw] = D_TAO_windowed[:,:,:,iw] - smooth(D_TAO_windowed[:,:,:,iw],smooth_N)
    D_TAO_full_lf[:,:,:,iw] = least_squares_spectrum_t_multi(D_TAO_full_hf1[:,:,:,iw], t, min_period=2, max_period=370, NSR=NSR, reconstruct_min_period = cutoff)[2]

D_TAO_full_hf = D_TAO_full_hf1 - D_TAO_full_lf
D_TAO_full_hf -= np.nanmean(D_TAO_full_hf,axis=0)    
```

```python
# Now do zonal averages carefully

# For full NEMO, do a standard zonal average
D_NEMO_full_za = np.nanmean(D_NEMO_full_hf, axis=2)

# For sampled NEMO and TAO, need to check there are more than one data points in the zonal average
D_NEMO_sampled_za = np.zeros((D_NEMO_sampled_hf.shape[0],D_NEMO_sampled_hf.shape[1]))*np.nan
D_TAO_year_za  = np.zeros((D_TAO_year_hf.shape[0],D_TAO_year_hf.shape[1]))*np.nan

for it in range(0,370):
    for iy in range(0,D_TAO_year_hf.shape[1]):
        if (np.sum(~np.isnan(D_TAO_year_hf[it,iy,:])) > 1):
            D_TAO_year_za[it,iy] = np.nanmean(D_TAO_year_hf[it,iy,:])
            D_NEMO_sampled_za[it,iy] = np.nanmean(D_NEMO_sampled_hf[it,iy,:])

D_TAO_full_za  = np.zeros((D_TAO_full_hf.shape[0],D_TAO_full_hf.shape[1],D_TAO_full_hf.shape[3]))*np.nan
for iw in range(0,Nwindows):
    for it in range(0,370):
        for iy in range(0,D_TAO_full_hf.shape[1]):
            if (np.sum(~np.isnan(D_TAO_full_hf[it,iy,:,iw])) > 1):
                D_TAO_full_za[it,iy,iw] = np.nanmean(D_TAO_full_hf[it,iy,:,iw])
```

```python
# Now do least squares fits:

freqs, power_NEMO_full, fit_NEMO_full = least_squares_spectrum_t_y(
    D_NEMO_full_za, t, lat_NEMO, min_period=2, max_period=370,y_modes=P_modes_NEMO,  NSR=NSR, max_period_cutoff = cutoff)

freqs, power_NEMO_sampled, fit_NEMO_sampled = least_squares_spectrum_t_y(
    D_NEMO_sampled_za, t, lat_TAO, min_period=2, max_period=370,y_modes=P_modes_TAO,  NSR=NSR, max_period_cutoff = cutoff)

freqs, power_TAO_year, fit_TAO_year = least_squares_spectrum_t_y(
    D_TAO_year_za, t, lat_TAO, min_period=2, max_period=370,y_modes=P_modes_TAO,  NSR=NSR, max_period_cutoff = cutoff)

freqs, power_TAO_full, fit_TAO_full = least_squares_spectrum_t_y(
    D_TAO_full_za, t, lat_TAO, min_period=2, max_period=370,y_modes=P_modes_TAO,  NSR=NSR, max_period_cutoff = cutoff)

power_TAO_full_av = np.nanmean(power_TAO_full,axis=2)
```

```python
fig, ax = plt.subplots(2,3,figsize = (15,10), constrained_layout=True)

ymin = 2e-3
ymax = 0.5
nfilt=5

for i in range(6):
    i1 = int(np.floor(i/3))
    i2 = i%3
    l0, = ax[i1,i2].plot(freqs,smooth(power_TAO_year[:,i],nfilt),'-', color='red',linewidth=2,zorder=6)
    l1, = ax[i1,i2].plot(freqs,smooth(power_NEMO_sampled[:,i],nfilt),'-', color='blue',linewidth=2,zorder=5)
    l2, = ax[i1,i2].plot(freqs,power_TAO_full_av[:,i],'-', color='black',linewidth=2,zorder=7)
    l3, = ax[i1,i2].plot(freqs,smooth(power_NEMO_full[:,i],nfilt),'-.', color='black',linewidth=1.5,zorder=4)
    l4, = ax[i1,i2].plot([freqbc1[i],freqbc1[i]],[ymin,ymax],'--',color='magenta',zorder=1)
    l5, = ax[i1,i2].plot([freqbc2[i],freqbc2[i]],[ymin,ymax],'--',color='gray',zorder=1)
    ax[i1,i2].text(0.2,0.6,f'Mode {i}',weight='bold',fontsize=14)
    ax[i1,i2].set_yscale('log')
    ax[i1,i2].set_xlim([0.05,0.4])
    ax[i1,i2].set_ylim([ymin,ymax])
    if i == 3:
        l0.set_label('TAO: sampled')
        l1.set_label('NEMO: sampled')
    if i == 4:
        l2.set_label('TAO: full data')
        l3.set_label('NEMO: full data')
    if i == 5:
        l4.set_label('BC1 freq')
        l5.set_label('BC2 freq')

ax[1,0].legend(facecolor='white',fontsize=14,framealpha=1,loc=3)
ax[1,1].legend(facecolor='white',fontsize=14,framealpha=1,loc=3)
ax[1,2].legend(facecolor='white',fontsize=14,framealpha=1,loc=3)
ax[0,1].axes.yaxis.set_ticklabels([])
ax[0,2].axes.yaxis.set_ticklabels([])
ax[1,1].axes.yaxis.set_ticklabels([])
ax[1,2].axes.yaxis.set_ticklabels([])
ax[0,0].axes.xaxis.set_ticklabels([])
ax[0,1].axes.xaxis.set_ticklabels([])
ax[0,2].axes.xaxis.set_ticklabels([])

ax[1,0].set_xlabel('Frequency (cpd)')
ax[1,1].set_xlabel('Frequency (cpd)')
ax[1,2].set_xlabel('Frequency (cpd)')
ax[0,0].set_ylabel('Power [cm]')
ax[1,0].set_ylabel('Power [cm]')

fig.savefig('../figures/figure2.png')
```
