## Meridional and temporal fit to TAO and NEMO data: zonal structure

```python
import sys
sys.path.append('../')
import matplotlib.pyplot as plt
import numpy as np
import xarray as xr
from numpy import pi
import scipy.io as sio
import matplotlib.colors as Colors
%matplotlib inline 
import warnings
from tqdm import tqdm
from tools.transform_tools import *
from tools.data_processing_tools import *
from tools.theoretical_tools import *

warnings.filterwarnings('ignore')
plt.rcParams.update({'font.size': 14, 'font.size': 14})
plt.rcParams['figure.figsize'] = (20, 10)
plt.rcParams['text.usetex'] = False

%load_ext autoreload
%autoreload 2
```

```python
smooth_N = 21 # Rolling mean length for first high pass filter
NSR = 35 # Noise to signal ratio for least squares fit
nmodes=6 # Number of meridional modes to fit to
cutoff = 20 # Cutoff period for second high pass filter
```

```python
# Load in full TAO dynamic height data 
t_TAO_full, lat_TAO, lon_TAO, lon_TAO_midpoints, D_TAO_full, ds_TAO_full = load_TAO(NEMO_year=False)

# Load in full NEMO dynamic height data (subsample in longitude to speed up these calculations).
t, lat_NEMO, lon_NEMO, D_NEMO_full, ds_NEMO_full = load_NEMO(daily_mean=True,lons='default',lats='default',lon_lims=[lon_TAO_midpoints[0],lon_TAO_midpoints[-1]])

# Load in year of TAO dynamic height data 
t, lat_TAO, lon_TAO, lon_TAO_midpoints, D_TAO_year, ds_TAO_year = load_TAO(NEMO_year=True)

# Load in sampled NEMO dynamic height data
t, lat_TAO, lon_TAO, D_NEMO_sampled, ds_NEMO_sampled = load_NEMO(daily_mean=True,lons=lon_TAO,lats=lat_TAO)

# Add in gaps from TAO data to NEMO dynamic height data:
D_NEMO_sampled = np.where(np.isnan(D_TAO_year),np.nan,D_NEMO_sampled)

```

```python
# Get the theoretical frequencies
freqbc1, freqbc2, c1, c2, lon_freqs = find_predicted_freqs(nmodes,lon_lims=[lon_TAO_midpoints[0],lon_TAO_midpoints[-1]],lat_lims=[-12,12],average_lon = False)
```

```python
# Load in meridional modes. Use the first baroclinic mode.
P_modes_NEMO = calc_meridional_modes(lat_NEMO,N=nmodes,cm=np.nanmean(c1))[1]
P_modes_TAO = calc_meridional_modes(lat_TAO,N=nmodes,cm=np.nanmean(c1))[1]
```

```python
# Do high pass filters (two stage approach - first remove a running mean to remove endpoint issues)

D_NEMO_full_hf1 = D_NEMO_full - smooth(D_NEMO_full,smooth_N)
D_NEMO_sampled_hf1 = D_NEMO_sampled - smooth(D_NEMO_sampled,smooth_N)
D_TAO_year_hf1 = D_TAO_year - smooth(D_TAO_year,smooth_N)

D_NEMO_full_lf = least_squares_spectrum_t_multi(D_NEMO_full_hf1, t, min_period=2, max_period=370, NSR=NSR, reconstruct_min_period = cutoff)[2]
D_NEMO_full_hf = D_NEMO_full_hf1 - D_NEMO_full_lf
D_NEMO_full_hf -= np.nanmean(D_NEMO_full_hf,axis=0)

D_NEMO_sampled_lf = least_squares_spectrum_t_multi(D_NEMO_sampled_hf1, t, min_period=2, max_period=370, NSR=NSR, reconstruct_min_period = cutoff)[2]
D_NEMO_sampled_hf = D_NEMO_sampled_hf1 - D_NEMO_sampled_lf
D_NEMO_sampled_hf -= np.nanmean(D_NEMO_sampled_hf,axis=0)

D_TAO_year_lf = least_squares_spectrum_t_multi(D_TAO_year_hf1, t, min_period=2, max_period=370, NSR=NSR, reconstruct_min_period = cutoff)[2]
D_TAO_year_hf = D_TAO_year_hf1 - D_TAO_year_lf
D_TAO_year_hf -= np.nanmean(D_TAO_year_hf,axis=0)
```

```python
# For the full TAO data, we need a windowed approach. Let's first create the windowed data, in yearly (370 day) intervals
window_width=370
Nsamples = D_TAO_full.shape[0]
window_starts = np.arange(0,Nsamples-window_width,window_width)
window_mid = window_starts+window_width/2
Nwindows = window_starts.shape[0]
D_TAO_windowed = np.zeros((window_width,D_TAO_full.shape[1],D_TAO_full.shape[2],Nwindows))
D_TAO_full_lf = np.zeros_like(D_TAO_windowed)
D_TAO_full_hf1 = np.zeros_like(D_TAO_windowed)
for iw in range(Nwindows):
    D_TAO_windowed[:,:,:,iw] = D_TAO_full[window_starts[iw]:window_starts[iw]+window_width]
    D_TAO_full_hf1[:,:,:,iw] = D_TAO_windowed[:,:,:,iw] - smooth(D_TAO_windowed[:,:,:,iw],smooth_N)
    D_TAO_full_lf[:,:,:,iw] = least_squares_spectrum_t_multi(D_TAO_full_hf1[:,:,:,iw], t, min_period=2, max_period=370, NSR=NSR, reconstruct_min_period = cutoff)[2]

D_TAO_full_hf = D_TAO_full_hf1 - D_TAO_full_lf
D_TAO_full_hf -= np.nanmean(D_TAO_full_hf,axis=0)    
```

```python
# Now do least squares fits:

freqs, power_NEMO_full, fit_NEMO_full = least_squares_spectrum_t_y(
    D_NEMO_full_hf, t, lat_NEMO, min_period=2, max_period=370,y_modes=P_modes_NEMO,  NSR=NSR, max_period_cutoff = cutoff)

freqs, power_NEMO_sampled, fit_NEMO_sampled = least_squares_spectrum_t_y(
    D_NEMO_sampled_hf, t, lat_TAO, min_period=2, max_period=370,y_modes=P_modes_TAO,  NSR=NSR, max_period_cutoff = cutoff)

freqs, power_TAO_year, fit_TAO_year = least_squares_spectrum_t_y(
    D_TAO_year_hf, t, lat_TAO, min_period=2, max_period=370,y_modes=P_modes_TAO,  NSR=NSR, max_period_cutoff = cutoff)

```

```python
# Least squares function can't take 4 dimensions, need a for loop for windows
power_TAO_full = np.zeros((freqs.shape[0],nmodes,D_TAO_full_hf.shape[2],Nwindows))
fit_TAO_full = np.zeros_like(D_TAO_full_hf)
for iw in range(Nwindows):
    freqs, power_TAO_full[:,:,:,iw], fit_TAO_full[:,:,:,iw] = least_squares_spectrum_t_y(
        D_TAO_full_hf[:,:,:,iw], t, lat_TAO, min_period=2, max_period=370,y_modes=P_modes_TAO,  NSR=NSR, max_period_cutoff = cutoff)

power_TAO_full_av = np.nanmean(power_TAO_full,axis=3)
```

```python
fig, ax = plt.subplots(4,6,figsize = (15,15),constrained_layout=True)
vmax = 0.4
vmin = 0.0
nfilt = 5

# Plot full NEMO
for i in range(ax.shape[1]):
    p = ax[0,i].pcolormesh(lon_NEMO, freqs,smooth(power_NEMO_full[:,i,:],nfilt),vmin=vmin,vmax=vmax)

# Plot spatially sampled NEMO
for i in range(ax.shape[1]):
    p = ax[1,i].pcolormesh(lon_TAO_midpoints, freqs,smooth(power_NEMO_sampled[:,i,:],nfilt),vmin=vmin,vmax=vmax)

# Plot spatially sampled TAO
for i in range(ax.shape[1]):
    p = ax[2,i].pcolormesh(lon_TAO_midpoints, freqs,smooth(power_TAO_year[:,i,:],nfilt),vmin=vmin,vmax=vmax)

# Plot full TAO
nfilt = 1
for i in range(ax.shape[1]):
    p = ax[3,i].pcolormesh(lon_TAO_midpoints, freqs,smooth(power_TAO_full_av[:,i,:],nfilt),vmin=vmin,vmax=vmax)

# Plot frequency curves
for i in range(ax.shape[0]):
    for j in range(ax.shape[1]):
        ax[i,j].plot(lon_freqs,freqbc1[j,:],'--',linewidth=2,color='red')
        ax[i,j].plot(lon_freqs,freqbc2[j,:],'--',linewidth=2,color='black')

# Set titles
for i in range(ax.shape[1]):
    ax[0,i].set_title(f'Mode {i}')

# Set ylims
for i in range(ax.shape[0]):
    for j in range(ax.shape[1]):
        ax[i,j].set_ylim([0.05,0.4])
        
# Set xlims
for i in range(ax.shape[0]):
    for j in range(ax.shape[1]):
        ax[i,j].set_xlim([lon_TAO_midpoints[0],lon_TAO_midpoints[-1]])
        

# Remove x ticklabels
for i in range(ax.shape[0]-1):
    for j in range(ax.shape[1]):
        ax[i,j].axes.xaxis.set_ticklabels([])

# Remove y ticklabels
for i in range(ax.shape[0]):
    for j in range(1,ax.shape[1]):
        ax[i,j].axes.yaxis.set_ticklabels([])
        
# Set colorbars
for i in range(ax.shape[0]):
    fig.colorbar(p, ax = ax[i,5])

ax[0,0].set_ylabel('NEMO: full data \nFrequency (cpd)',fontsize=16)
ax[1,0].set_ylabel('NEMO: sampled data \nFrequency (cpd)',fontsize=16)
ax[2,0].set_ylabel('TAO: sampled data \nFrequency (cpd)',fontsize=16)
ax[3,0].set_ylabel('TAO: full data \nFrequency (cpd)',fontsize=16)

fig.text(0.5, -0.02, 'Longitude ($^\circ$E)', ha='center',fontsize=16)
fig.savefig('../figures/figure3.png',transparent=False, bbox_inches="tight")
```
